```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
```

```python
column_1 = crime_ind.ix[:,0]

crime_indt=pd.DataFrame({"year": column_1.dt.year,
              "month": column_1.dt.month,
              "day": column_1.dt.day,
              "hour": column_1.dt.hour,
              "dayofyear": column_1.dt.dayofyear,
              "week": column_1.dt.week,
              "weekofyear": column_1.dt.weekofyear,
              "dayofweek": column_1.dt.dayofweek,
              "weekday": column_1.dt.weekday,
              "quarter": column_1.dt.quarter,
             })
```

```python
crime_indse=crime_ind.drop('timestamp',axis=1)
```

```python
crime_indd=pd.concat([crime_indt,crime_indse],axis=1)
```

```python
crime_indd.dropna(inplace=True)
```

```python
crime_indd
```

```python
X=crime_indd.iloc[:,[1,2,3,4,6,16,17]].values
```

```python
y=crime_indd.iloc[:,[10,11,12,13,14,15]].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=50)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,y_train)
```

```python
pred=knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,pred))
```

```python
knn.score(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

```python
error_rate = []
for i in range(1,100):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,100),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='yellow', markersize=7)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
prediction=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
dtree.score(X_test,y_test)
```

```python
dtree.score(X_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_pred=rfc.predict(X_test)
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, rfc_pred))
```
