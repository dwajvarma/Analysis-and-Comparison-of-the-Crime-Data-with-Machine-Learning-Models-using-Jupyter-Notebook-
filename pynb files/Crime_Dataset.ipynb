```python

crime_boolean=top_crime[(top_crime['Primary Type']=='BATTERY')|(top_crime['Primary Type']=='THEFT')]
crime_boolean
```

```python
crime_boolean=top_crime[['Primary Type','Arrest','Domestic']]
crime_boolean
```

```python
crime_boolean[['Arrest', 'Domestic']] = crime_boolean[['Arrest', 'Domestic']].astype(int)
crime_boolean
```

```python
crime_boolean1=crime_boolean[(crime_boolean['Primary Type']=='THEFT')|(crime['Primary Type']=='BATTERY')]
crime_boolean1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = crime_boolean1.drop('Primary Type',axis=1)
y = crime_boolean1['Primary Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30,random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictions=dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10)
rfc.fit(X_train, y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
print(classification_report(y_test,rfc_pred))
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, rfc_pred))
```

```python
import folium
```

```python
crime_map = folium.Map(location=[41.891398, -87.744384],zoom_start=11)
```

```python
c_map_district = cm.groupby('District').first().loc[:, ['Latitude', 'Longitude', 'Location Description', 'Arrest']]
```

```python
for i in range(len(c_map_district)):
    lat = c_map_district.iloc[i][0]
    long = c_map_district.iloc[i][1]
    popup_text = """District : {}<br>
                Arrest : {}<br>
                Location Description : {}<br>"""
    popup_text = popup_text.format(c_map_district.index[i],c_map_district.iloc[i][-1],c_map_district.iloc[i][-2])
    folium.CircleMarker(location = [lat, long], popup= popup_text).add_to(crime_map)
```
